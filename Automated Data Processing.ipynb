```python
import pandas as pd

# Load Excel data
file_path = "desktop/sample_sales_data.xlsx"
data = pd.read_excel(file_path)

# Display the first few rows of the data
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
cleaned_data = data.dropna()
cleaned_data = cleaned_data.rename(columns={"Unit sold": "unit_sold", "Total Revenue": "total_revenue"})
```

```python
# Example: Calculate total sales by region
sales_by_region = cleaned_data.groupby("Region")["total_revenue"].sum()

# Display the result
print(sales_by_region)
```

```python
# Save the processed data to a new Excel file
cleaned_data.to_excel("processed_sales_data.xlsx", index=False)
```

```python
import pandas as pd
```

```python
total_revenue = cleaned_data["total_revenue"].sum()
print ("done")
```

```python
cleaned_data['revenue_percentage'] = (cleaned_data['total_revenue'] / total_revenue) * 100
print ("done")
```

```python
# Find the region with the highest and lowest revenue
max_revenue_region = cleaned_data['total_revenue'].idxmax()
min_revenue_region = cleaned_data['total_revenue'].idxmin()
print ("done")
```

```python
print("Total Revenue: ", total_revenue)
print("Percentage contribution by each region:\n", cleaned_data[['total_revenue', 'revenue_percentage']])
print("Region with highest revenue: ", max_revenue_region)
print("Region with lowest revenue: ", min_revenue_region)
```

```python
import matplotlib.pyplot as plt
print ("done")
```

```python
# Bar chart for total revenue by region
plt.figure(figsize=(7, 4))
plt.bar(cleaned_data.index, cleaned_data['total_revenue'], color='skyblue')
plt.title('Total Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# Pie chart for percentage contribution by region
plt.figure(figsize=(6, 6))
plt.pie(cleaned_data['revenue_percentage'], labels=cleaned_data.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Revenue Percentage by Region')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```
